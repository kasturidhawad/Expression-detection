```python
import cv2
from deepface import DeepFace
import pandas as pd # for data storage
import time #tracking the duration of the real-time
#video processing.

# Create an empty DataFrame to store the emotions
columns = ['Frame', 'Emotion']
emotion_df = pd.DataFrame(columns=columns)
#This DataFrame will be used to store frame numbers and corresponding 
#emotions detected during real-time video processing.

faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(1)
if not cap.isOpened():
    cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise IOError("Cannot open webcam")

frame_number = 0
start_time = time.time()
end_time = start_time + 30  # Run for 30 seconds

#Enter a loop that continues as long as the current
#time is less than the end_time. In each iteration,
#it captures a frame from the webcam using cap.read() 
#and stores it in the frame variable.
while time.time() < end_time:
    ret, frame = cap.read()

    # Check if the frame was read correctly
    if not ret:
        break

    # Set enforce_detection to False to skip face detection
    result = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)
#Convert the captured frame to grayscale, as face detection typically works on grayscale images.
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    font = cv2.FONT_HERSHEY_SIMPLEX

    # Get the dominant emotion
    dominant_emotion = result[0]['dominant_emotion']

    # Append the frame number and emotion to the DataFrame
    emotion_df = emotion_df.append({'Frame': frame_number, 'Emotion': dominant_emotion}, ignore_index=True)

    cv2.putText(frame,
                dominant_emotion,
                (50, 50),
                font, 3,
                (0, 0, 255),
                2,
                cv2.LINE_4)

    cv2.imshow('Original video', frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

    frame_number += 1

cap.release()
cv2.destroyAllWindows()

# Now you have a DataFrame containing frame numbers and corresponding emotions
print(emotion_df)
```
